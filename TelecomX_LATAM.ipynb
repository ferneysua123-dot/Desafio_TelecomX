```python
# Cambio de los títulos de las Columnas (Viejo: Nuevo)
nuevas_columnas = {
    'customerID': 'ID_Cliente',
    'Churn': 'Abandono',
    'customer.gender': 'Genero',
    'customer.SeniorCitizen': 'Adulto_Mayor',
    'customer.Partner': 'Tiene_Pareja',
    'customer.Dependents': 'Tiene_Dependientes',
    'customer.tenure': 'Meses_Permanencia',
    'phone.PhoneService': 'Servicio_Telefonico',
    'phone.MultipleLines': 'Varias_Lineas',
    'internet.InternetService': 'Tipo_Internet',
    'internet.OnlineSecurity': 'Seguridad_Online',
    'internet.OnlineBackup': 'Respaldo_Online',
    'internet.DeviceProtection': 'Proteccion_Dispositivo',
    'internet.TechSupport': 'Soporte_Tecnico',
    'internet.StreamingTV': 'Streaming_TV',
    'internet.StreamingMovies': 'Streaming_Peliculas',
    'account.Contract': 'Tipo_Contrato',
    'account.PaperlessBilling': 'Factura_Digital',
    'account.PaymentMethod': 'Metodo_Pago',
    'account.Charges.Monthly': 'Cargo_Mensual',
    'account.Charges.Total': 'Cargo_Total'
}

df.rename(columns=nuevas_columnas, inplace=True)

```

```python
traduccion_valores = {
    "Yes": "Si",
    "No": "No",
    "Female": "Mujer",
    "Male": "Hombre",
    "No phone service": "Sin servicio telefonico",
    "No internet service": "Sin servicio de internet",
    "DSL": "DSL",
    "Fiber optic": "Fibra optica",
    "Month-to-month": "Mes a mes",
    "One year": "Un año",
    "Two year": "Dos años",
    "Electronic check": "Cheque electronico",
    "Mailed check": "Cheque por correo",
    "Credit card (automatic)": "Tarjeta de credito (automatica)",
    "Bank transfer (automatic)": "Transferencia bancaria (automatica)"
}
df.rename(columns=traduccion_valores, inplace=True)
# Aplicar traducciones a todas las columnas de tipo "object"
for col in df.select_dtypes(include=["object"]).columns:
    df[col] = df[col].replace(traduccion_valores)

print("Traducción aplicada. Vista previa:")
print(df.head(3))
```

```python
from numpy import int64
import pandas as pd

# Convertir 'Cargo_Total' a número
# 'coerce' convierte los espacios en blanco en NaN (nulos) para que no falle
df['Cargo_Total'] = pd.to_numeric(df['Cargo_Total'], errors='coerce')

# Llenar los valores nulos que aparecieron en 'Cargo_Total'
# Normalmente son clientes con 0 meses de permanencia
df['Cargo_Total'] = df['Cargo_Total'].fillna(0)

# Revisar cómo quedaron los tipos de datos
print(df.info())
```

```python
#Diagnóstico preventivo
print("Valores antes de limpiar:", df['Abandono'].unique())

#Limpieza de TEXTO
columnas_binarias = [
    'Abandono', 'Tiene_Pareja', 'Tiene_Dependientes', 'Varias_Lineas',
    'Servicio_Telefonico', 'Seguridad_Online', 'Respaldo_Online',
    'Proteccion_Dispositivo', 'Soporte_Tecnico', 'Streaming_TV',
    'Streaming_Peliculas', 'Factura_Digital'
]

for col in columnas_binarias:
    # Convertimos a string, quitamos espacios y pasamos a MINÚSCULAS
    df[col] = df[col].astype(str).str.strip().str.lower()

# 4. Mapeo SEGURO (fíjate que las llaves están en minúsculas)
mapeo = {
    'yes': 1,
    'si': 1,
    'no': 0,
    '': 0 # Explicitly map empty strings to 0 for 'Abandono' and other binary columns if they occur
}

# Usamos REPLACE para no destruir los datos si no hay coincidencia
df[columnas_binarias] = df[columnas_binarias].replace(mapeo)

# 5. Verificación final
print(df[columnas_binarias].head())
```

```python
df.info()
```

Ahora que los datos están limpios, es momento de crear la columna "Cuentas_Diarias". Utiliza la facturación mensual para calcular el valor diario, proporcionando una visión más detallada del comportamiento de los clientes a lo largo del tiempo.

```python
# Dividimos el cargo mensual por 30 (promedio de días al mes)
df['Cuentas_Diarias'] = df['Cargo_Mensual'] / 30
df['Cuentas_Diarias'] = df['Cuentas_Diarias'].round(2)
print(df[['ID_Cliente', 'Cargo_Mensual', 'Cuentas_Diarias']].head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Esto convierte '0' en 0 y '1' en 1
df['Abandono'] = pd.to_numeric(df['Abandono'], errors='coerce').fillna(0).astype(int)

plt.figure(figsize=(14, 5))

# --- GRÁFICO 1: Cantidad de Clientes (Barras) ---
plt.subplot(1, 2, 1)
# Arreglamos el FutureWarning usando hue='Abandono'
ax = sns.countplot(x='Abandono', data=df, hue='Abandono', palette='viridis', legend=False)
plt.title('Distribución de Evasión (Churn)')
plt.xticks([0, 1], ['Permanecieron (No)', 'Se Fueron (Yes)'])
plt.ylabel('Número de Clientes')

for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha = 'center', va = 'center', xytext = (0, 5), textcoords = 'offset points')

# --- GRÁFICO 2: Proporción de Clientes (Pastel) ---
plt.subplot(1, 2, 2)

churn_counts = df['Abandono'].value_counts().sort_index()
labels_map = {0: 'Permanecieron', 1: 'Se Fueron'}
colors_map = {0: '#45B39D', 1: '#EC7063'}

# Generamos listas basadas en lo que realmente existe en los datos
current_labels = [labels_map[idx] for idx in churn_counts.index]
current_colors = [colors_map[idx] for idx in churn_counts.index]
current_explode = [0.1 if idx == 1 else 0 for idx in churn_counts.index]

# Si solo hay una categoría, el explode debe ser de un solo elemento [0]
if len(current_explode) != len(churn_counts):
    current_explode = [0] * len(churn_counts)

plt.pie(
    churn_counts,
    autopct='%1.1f%%',
    colors=current_colors,
    labels=current_labels,
    startangle=90,
    explode=current_explode
)
plt.title('Proporción de Evasión (%)')

plt.tight_layout()
plt.show()
```

Ahora, exploraremos cómo se distribuye la evasión según variables categóricas, como género, tipo de contrato, método de pago, entre otras.

Este análisis puede revelar patrones interesantes, por ejemplo, si los clientes de ciertos perfiles tienen una mayor tendencia a cancelar el servicio, lo que ayudará a orientar acciones estratégicas

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Definimos las variables categóricas que queremos explorar
variables_interes = ['Tipo_Contrato', 'Metodo_Pago', 'Genero']

# 2. Configuramos la figura (3 gráficos en una fila)
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(20, 6))

for i, col in enumerate(variables_interes):
    # Creamos un gráfico de barras que cuente los clientes y los divida por 'Abandono'
    sns.countplot(x=col, hue='Abandono', data=df, ax=axes[i], palette='magma')

    axes[i].set_title(f'Abandono por {col}')
    axes[i].set_xticklabels(axes[i].get_xticklabels(), rotation=45)
    axes[i].set_xlabel('')
    axes[i].legend(title='Abandono', labels=['No', 'Sí'])

plt.tight_layout()
plt.show()

# 3. Análisis numérico rápido: Tasa de abandono por Contrato
print("Tasa de Abandono por Tipo de Contrato (%):")
print((df.groupby('Tipo_Contrato')['Abandono'].mean() * 100).round(2))
```

```python
# 1. Configuramos la figura para ver dos comparaciones clave
plt.figure(figsize=(16, 6))

# --- GRÁFICO 1: Meses de Permanencia vs Abandono ---
# Nos dice si la antigüedad influye en la fuga
plt.subplot(1, 2, 1)
sns.boxplot(x='Abandono', y='Meses_Permanencia', data=df, hue='Abandono', palette='magma', legend=False)
plt.title('Distribución de Antigüedad por Abandono')
plt.xticks([0, 1], ['Se Quedan', 'Se Van'])
plt.xlabel('Estado del Cliente')
plt.ylabel('Meses de Permanencia')

# --- GRÁFICO 2: Cargo Mensual vs Abandono ---
# Nos dice si los precios altos empujan a la gente a irse
plt.subplot(1, 2, 2)
sns.kdeplot(df[df['Abandono'] == 0]['Cargo_Mensual'], label='Se Quedan', fill=True, color='#45B39D')
sns.kdeplot(df[df['Abandono'] == 1]['Cargo_Mensual'], label='Se Van', fill=True, color='#EC7063')
plt.title('Densidad de Cargos Mensuales: ¿Quién paga más?')
plt.xlabel('Cargo Mensual ($)')
plt.ylabel('Densidad')
plt.legend()

plt.tight_layout()
plt.show()

# 3. Resumen numérico para confirmar los gráficos
print("Promedios por grupo de Abandono:")
print(df.groupby('Abandono')[['Meses_Permanencia', 'Cargo_Mensual', 'Cuentas_Diarias']].mean().round(2))
```
